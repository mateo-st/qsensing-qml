```python
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd
```

### Plotting Accuracy for Machine Learning on classical and on quantum data

Accuracy $:= \frac{\text{# Correct classifications}}{\text{# Total classifications}}$

```python
# Data from simulations

validation_acc_qdata = {1: [0.5, 0.6, 0.7, 0.6, 0.45], 2: [0.65, 0.7, 0.6, 0.8, 0.4], 3: [0.65, 0.75, 0.45, 0.65, 0.7], 4: [0.7, 0.55, 0.75, 0.35, 0.5], 5: [0.75, 0.7, 0.8, 0.65, 0.75], 6: [0.7, 0.7, 0.85, 0.7, 0.5]}

validation_acc_cdata = {'1': [0.425, 0.525, 0.525, 0.575, 0.55], '2': [0.5, 0.5, 0.55, 0.575, 0.475], '3': [0.55, 0.45, 0.525, 0.5, 0.5], '4': [0.525, 0.475, 0.525, 0.6, 0.45], '5': [0.55, 0.475, 0.525, 0.575, 0.525], '6': [0.6, 0.45, 0.475, 0.625, 0.45]}


```

```python
# reshape and labe data


dics = [validation_acc_qdata, validation_acc_cdata]
names = ["Quantum data", "Classical data"]

rows = []
for i, dic in enumerate(dics):
    for depth in dic.keys():
        for acc in dic[depth]:
            rows.append([depth, acc, names[i]])
            
df = pd.DataFrame(data = rows, columns = ["Circuit Depth", "Accuracy", "Data Type"])

```

```python
sns.lineplot(data=df, x="Circuit Depth", y="Accuracy", hue = "Data Type", estimator='max')
plt.xlim([1, 6])
plt.ylim([0.3, 1])

```
